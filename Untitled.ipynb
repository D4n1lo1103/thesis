```python
# Percorso del file (modifica se necessario)
json_path = "../export_total_db.json"

with open(json_path, encoding="utf-8") as f:
    data = json.load(f)

df = pd.DataFrame(data["rows"])

df.head()
df.info()
```

Questo blocco realizza l’acquisizione del dataset da un file in formato JSON e la sua trasformazione in un DataFrame, struttura dati che consente una rappresentazione tabellare idonea alle successive fasi di analisi esplorativa, pulizia e modellazione. La lettura avviene mediante parsing del JSON in un oggetto Python (tipicamente un dizionario), dal quale viene estratta la lista di record associata alla chiave rows.

La costruzione del DataFrame consente di interpretare ciascun record come un’osservazione e ciascun attributo come una variabile. Le prime ispezioni (head() e info()) hanno funzione diagnostica: permettono di verificare la correttezza del caricamento, la dimensione del campione, i tipi di dato e la presenza di valori mancanti, fornendo elementi oggettivi per pianificare il preprocessing.

```python
# Conversione robusta delle colonne data (valori non parseabili -> NaT)
date_cols = ["Data_Vendita", "Dettaglio_Data_Sbarco"]
for c in date_cols:
    if c in df.columns:
        df[c] = pd.to_datetime(df[c], errors="coerce")

# Selezione delle variabili rilevanti (mantiene solo le colonne presenti nel dataset)
selected_features = [
    "Data_Vendita",
    "Modalita_Acquisto",
    "Peso_Netto",
    "Peso_Lordo",
    "Prezzo_Unitario",
    "Prezzo_Totale",
    "Codice_Attrezzatura",
    "ZonaFao_GSA_Codice",
    "Codice_Specie_FAO",
    "Dettaglio_Data_Sbarco",
    "Dettaglio_Qualita"
]
df = df[[c for c in selected_features if c in df.columns]].copy()
```

Questo blocco ha una duplice finalità: (i) normalizzare le variabili temporali e (ii) restringere il dataset a un insieme di colonne ritenute informative per le successive analisi. Le colonne contenenti date vengono convertite in formato datetime, condizione necessaria per eseguire operazioni temporali (ad esempio estrazione di anno/mese/giorno della settimana e aggregazioni per periodo). L’opzione errors="coerce" implementa una conversione robusta: valori non conformi vengono trasformati in NaT, evitando interruzioni dell’esecuzione e rendendo esplicita la presenza di dati temporali non validi.

Successivamente viene applicata una selezione esplicita delle variabili. Tale operazione riduce la dimensionalità, migliora la tracciabilità metodologica e consente di focalizzare l’analisi su variabili coerenti con gli obiettivi: informazioni economiche (prezzi), quantitative (pesi), temporali (vendita/sbarco) e categoriali (zona FAO/GSA, attrezzatura, qualità e specie). La comprensione if c in df.columns garantisce la robustezza dell’esecuzione anche qualora alcune colonne non fossero presenti nel file sorgente.

```python
# Calcolo della percentuale di valori mancanti per ciascuna variabile
missing_percentage = df.isnull().mean() * 100

# Selezione delle variabili con percentuale di missing superiore al 50%
missing_over_50 = (
    missing_percentage[missing_percentage > 50]
    .sort_values(ascending=False)
)

print(missing_over_50)
```

Questo blocco esegue un’analisi quantitativa della presenza di valori mancanti all’interno del dataset. In particolare, per ciascuna variabile viene calcolata la percentuale di osservazioni prive di valore, normalizzando il conteggio dei valori nulli rispetto alla numerosità complessiva del campione.

La selezione delle sole variabili con una percentuale di missing superiore al 50% ha finalità diagnostica: consente di individuare attributi caratterizzati da una scarsa disponibilità informativa, che potrebbero risultare poco affidabili o richiedere strategie di imputazione aggressive. In questa fase tali variabili non vengono automaticamente eliminate, ma l’analisi fornisce una base oggettiva per le decisioni di preprocessing e feature selection adottate nei passaggi successivi.

```python
# Rimozione dei record privi della variabile economica principale
df = df.dropna(subset=["Prezzo_Unitario"]).copy()

# Eliminazione di prezzi unitari negativi
df = df[df["Prezzo_Unitario"] >= 0]

# Eliminazione di pesi negativi (i valori mancanti vengono mantenuti)
if "Peso_Netto" in df.columns:
    df = df[df["Peso_Netto"].isna() | (df["Peso_Netto"] >= 0)]

if "Peso_Lordo" in df.columns:
    df = df[df["Peso_Lordo"].isna() | (df["Peso_Lordo"] >= 0)]

# Controllo di coerenza fisica: Peso_Lordo >= Peso_Netto
if "Peso_Lordo" in df.columns and "Peso_Netto" in df.columns:
    df = df[
        df["Peso_Lordo"].isna()
        | df["Peso_Netto"].isna()
        | (df["Peso_Lordo"] >= df["Peso_Netto"])
    ].copy()
```

Questo blocco implementa una fase di pulizia dei dati basata su vincoli di coerenza semantica e fisica delle variabili. In primo luogo, vengono rimossi i record privi del valore di Prezzo_Unitario, in quanto tale variabile rappresenta l’informazione economica centrale dell’analisi e la sua assenza renderebbe l’osservazione non interpretabile.

Successivamente, vengono eliminate tutte le osservazioni che presentano prezzi unitari negativi, poiché incompatibili con il significato economico della variabile. In modo analogo, vengono rimossi i record con pesi negativi, mantenendo invece i valori mancanti, che verranno gestiti in una fase successiva tramite imputazione.

Infine, viene imposto un vincolo di coerenza fisica tra Peso_Lordo e Peso_Netto, richiedendo che il peso lordo sia maggiore o uguale al peso netto ogniqualvolta entrambi siano presenti. L’applicazione di tali controlli consente di garantire che il dataset risultante sia coerente con il dominio applicativo e riduce il rischio che osservazioni errate influenzino le analisi esplorative e l’addestramento del modello.

```python
# Selezione delle variabili numeriche di interesse
num_cols_corr = [
    c for c in ["Peso_Netto", "Peso_Lordo", "Prezzo_Unitario", "Prezzo_Totale"]
    if c in df.columns
]

# Calcolo della matrice di correlazione
corr = df[num_cols_corr].corr()

# Visualizzazione tramite heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Matrice di Correlazione tra Variabili Numeriche")
plt.show()
```

Questo blocco analizza le relazioni lineari tra le principali variabili numeriche mediante il coefficiente di correlazione di Pearson. La matrice di correlazione consente di individuare legami positivi o negativi tra le variabili e di valutarne l’intensità. Tale analisi è particolarmente rilevante nella fase di preprocessing, in quanto la presenza di correlazioni molto elevate può indicare ridondanza informativa e potenziali problemi di multicollinearità, che possono influenzare negativamente la stabilità e l’interpretabilità dei modelli predittivi.

```python
# Eliminazione di Peso_Lordo se fortemente correlato a Peso_Netto
if "Peso_Lordo" in df.columns and "Peso_Netto" in df.columns:
    df = df.drop(columns=["Peso_Lordo"]).copy()
```

In seguito all’analisi delle correlazioni, la variabile Peso_Lordo viene rimossa dal dataset in quanto fortemente correlata con Peso_Netto. L’eliminazione di una delle due variabili riduce la multicollinearità e semplifica la struttura dei dati, evitando che informazioni sostanzialmente ridondanti influenzino l’addestramento del modello. Tale scelta è coerente con le buone pratiche di feature selection e contribuisce a migliorare la robustezza della modellazione successiva.
